```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/Salaries.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
num_rows=df.shape[0]
num_columns=df.shape[1]
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
print("Data types of each column:")
print(df.dtypes)
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
totalpay_stats = df['TotalPay'].describe()
```

```python
# Calculate basic statistics for 'TotalPay'
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

# Determine the range of salaries
salary_range = max_salary - min_salary

# Find the standard deviation of salaries
std_dev_salary = df['TotalPay'].std()
```

```python
print('mean salary=',mean_salary)
print('median salary=',median_salary)
print('mode salary=',mode_salary)
print('min salary=',min_salary)
print('max salary=',max_salary)
print('salary range=',salary_range)
print('std dev salary=',std_dev_salary)
```

Handle missing data in the 'TotalPay' column by replacing missing values with the mean.
We use this method because the 'TotalPay' column is numerical,
and using the mean helps to maintain the overall distribution of the data
using inplace=True modifies the original object directly

```python
df['TotalPay'].fillna(df['TotalPay'].mean(), inplace=True)
```

```python
plt.hist(df['TotalPay'], bins=30, edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
department_proportions = df['JobTitle'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(department_proportions, labels=department_proportions.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
# Group data by 'JobTitle' and calculate summary statistics
grouped_data = df.groupby('JobTitle')['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])

# Display the grouped data
print("Grouped Analysis by JobTitle:")
print(grouped_data)
```

```python
# Correlation analysis between 'TotalPay' and 'OvertimePay'
correlation = df['TotalPay'].corr(df['OvertimePay'])
print("\nCorrelation between 'TotalPay' and 'OvertimePay':", correlation)
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OvertimePay', y='TotalPay', data=df)
plt.title(f'Scatter Plot: TotalPay vs OvertimePay\nCorrelation: {correlation:.2f}')
plt.xlabel('OvertimePay')
plt.ylabel('TotalPay')
plt.show()
```

# 5-Group Analysis:

The provided code conducted an analysis on the dataset based on different job titles (JobTitle). First, it grouped the data according to the job titles, then calculated various summary statistics for salaries within each job title category. These statistics included the total count, mean, median, minimum, and maximum salary values. The results were presented in a new DataFrame, showcasing these statistics for each job title.

This analysis helps us understand how employee salaries vary across a diverse set of job titles. We can observe differences in averages, medians, as well as the minimum and maximum salaries, providing valuable insights for decision-making based on these statistical measures.

# 6-Simple Correlation Analysis:

Correlation analysis between 'TotalPay' and 'OvertimePay'
Correlation Analysis:

Provide us with an understanding of how additional salaries affect overall salaries and the extent of the relationship between them.

Scatter Plot:

It visually shows how the data is distributed and whether there is a trend or correlation between total salaries and additional salaries.

```python
correlation = df['TotalPay'].corr(df['OvertimePay'])
print("\nCorrelation between 'TotalPay' and 'OvertimePay':", correlation)
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OvertimePay', y='TotalPay', data=df)
plt.title(f'Scatter Plot: TotalPay vs OvertimePay\nCorrelation: {correlation:.2f}')
plt.xlabel('OvertimePay')
plt.ylabel('TotalPay')
plt.show()
```
